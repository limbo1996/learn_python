```python
# os.path模块
# os.path模块包含很多文件名与文件路径相关的有用函数
## 处理绝对路径个相对路径
```

```python
# os.path.abspath()返回绝对路径，将相对路径换成绝对路径
os.path.abspath('.')
```

```python
os.path.abspath('.\\Scripts')
```

```python
# 调用os.path.isabs(), 如果参数是一个绝对路径， 返回True, 否则为False
```

```python
os.path.isabs('.')
```

```python
os.path.isabs(os.path.abspath('.'))
```

```python
# os.path.basename得到最后一个斜杠之后的内容
# os.path.dirname得到最后一个斜杠之前的内容
```

```python
path = 'C:\\Windows\\system32\\calc.exe'
```

```python
os.path.basename(path)
```

```python
os.path.dirname(path)
```

```python
# os.path.split()可以分割路径得到两个结果放在元组中，分别是路径和文件名
calcFilePath = 'C:\\Windows\\System32\\calc.exe'
os.path.split(calcFilePath)
```
